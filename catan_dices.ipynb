# Simulate throwing two(Or N) dice for Lacs of times, find their sums, and count the frequency of each sum.

```python
import random
from collections import Counter
import matplotlib.pyplot as plt
```

```python
def dice_sum_experiment():
    DICES = 7  # Number of dice used in the experiment, typically 2 in Catan; Its fun to increase this number!
    dice_sum_result = 0
    for _ in range(DICES):
        dice_sum_result += random.randint(1, 6)
    return dice_sum_result
```

```python
TRIALS = 900_000  # Nine Lacs
storing_result = []
for _ in range(TRIALS):
    result_of_exp = dice_sum_experiment()
    storing_result.append(result_of_exp)
# print(storing_result)
# Example Output for when two dices are thrown.: [6, 5, 6, 5, 8, 4, 2, 11, 8, 7, 11, 8, 8, 7, 6, 4, 4, 4, 6, 2] 
```

```python
# Counting
counting_sums = Counter(storing_result)
print(counting_sums)
```

```python
# Sorting as per their counts
sorting_counted_sums = sorted(counting_sums.items(), key = lambda x : x[1])
print(sorting_counted_sums)
```

```python
# Extract the sums and their counts
sums = [item[0] for item in sorting_counted_sums]
counts = [item[1] for item in sorting_counted_sums]

# Plotting the data
plt.figure(figsize=(10, 6))
plt.bar(sums, counts, color='blue')
plt.xlabel('Sum of Dice Rolls')
plt.ylabel('Frequency')
plt.title('Frequency of Sums from Dice Rolls')
plt.xticks(sums) 
plt.grid(axis='y')
plt.show()
```
